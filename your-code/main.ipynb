## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have

```python
# your code here
# Generate 1000 uniform random numbers between 0 and 10
data_numpy = np.random.uniform(low=0, high=10, size=1000)

# Plot histogram
plt.hist(data_numpy, bins=30, color='skyblue', edgecolor='black')
plt.title('Uniform Distribution using NumPy')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

**Your task:**

1. Based on the code above, write a function that generates uniformly distributed random numbers. There are several requirements for your function:
    * It should accept 3 parameters: 
        * `bottom` - the lower boundary of the generated numbers
        * `ceiling` - the upper boundary of the generated numbers
        * `count` - how many numbers to generate
    * It should return an array of uniformly distributed random numbers

1. Call your function with 2 sets of params below:
    * bottom=10, ceiling=15, count=100
    * bottom=10, ceiling=60, count=1,000

1. Plot the uniform distributions generated above using histograms, where x axis is the value and y axis is the count. Let the histogram's number of bins be 10.

Your output should look like below:

![uniform distribution](ud.png)

```python
# your code here
def generate_uniform(bottom, ceiling, count):
    """
    Generates uniformly distributed random numbers.
    
    Parameters:
        bottom (float): lower boundary
        ceiling (float): upper boundary
        count (int): number of samples to generate
    
    Returns:
        np.array: uniformly distributed random numbers
    """
    return np.random.uniform(low=bottom, high=ceiling, size=count)
```

```python
# your code here
data1 = generate_uniform(10, 15, 100)
data2 = generate_uniform(10, 60, 1000)
```

```python
# your code here
plt.figure(figsize=(8,4))

plt.subplot(1, 2, 1)
plt.hist(data1, bins=10, color='steelblue', edgecolor='black')
plt.title('Uniform (10–15), n=100')
plt.xlabel('Value')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
plt.hist(data2, bins=10, color='steelblue', edgecolor='black')
plt.title('Uniform (10–60), n=1000')
plt.xlabel('Value')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

How are the two distributions different?

```python
# your answer below 
# Both are uniformly distributed, but the larger range and higher count make the second histogram appear flatter and more consistent across bins.
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
from scipy.stats import norm 
```

```python
# your code here
def generate_normal(mean, std, count):
    """
    Generates normally distributed random numbers.
    
    Parameters:
        mean (float): the average (μ)
        std (float): the standard deviation (σ)
        count (int): number of samples to generate
        
    Returns:
        np.array: normally distributed random numbers
    """
    return np.random.normal(loc=mean, scale=std, size=count)
```

```python
# your code here
data1 = generate_normal(mean=10, std=1, count=1000)
```

```python
# your code here
data2 = generate_normal(mean=10, std=50, count=1000)
```

```python
# your code here
plt.figure(figsize=(8,4))

plt.subplot(1, 2, 1)
plt.hist(data1, bins=50, color='cornflowerblue', edgecolor='black')
plt.title('Normal(μ=10, σ=1)')
plt.xlabel('Value')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
plt.hist(data2, bins=50, color='lightcoral', edgecolor='black')
plt.title('Normal(μ=10, σ=50)')
plt.xlabel('Value')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
# your code here
import numpy as np
import matplotlib.pyplot as plt

# 1 Define a function to generate exponential numbers
def generate_exponential(mean, size):
    """
    Generates exponentially distributed random numbers.

    Parameters:
        mean (float): mean (λ⁻¹) of the distribution
        size (int): number of random samples to generate

    Returns:
        np.array: exponentially distributed random numbers
    """
    return np.random.exponential(scale=mean, size=size)

# 2 Generate two datasets
data1 = generate_exponential(mean=10, size=10)
data2 = generate_exponential(mean=10, size=100)

# 3 Plot histograms
plt.figure(figsize=(8,4))

plt.subplot(1, 2, 1)
plt.hist(data1, bins=100, color='cornflowerblue', edgecolor='black')
plt.title('Exponential Distribution (n=10)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(data2, bins=100, color='lightcoral', edgecolor='black')
plt.title('Exponential Distribution (n=100)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

How are the two distributions different?

Both come from the same exponential distribution (mean = 10), but with a larger sample size, the distribution more closely matches the theoretical exponential curve.

The mean changes, so the distribution changes as well. 

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# your code here
import math

# Given parameters
mean = 10        # mean time in minutes
lmbda = 1 / mean # rate parameter λ
x = 15           # time in minutes

# Exponential CDF formula: P(X < x) = 1 - e^(-λx)
probability = 1 - math.exp(-lmbda * x)

print(f"The probability that a customer spends less than 15 minutes in the bank is {probability:.4f}")
```
